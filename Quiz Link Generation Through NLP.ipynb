# AutoTokenizer & TFAutoModelForSequenceClassification

```python
import pandas as pd
import json
from transformers import pipeline, AutoTokenizer, TFAutoModelForSequenceClassification

# Load the quiz data from the JSON file
file_path = "quiz.json"
with open(file_path, "r") as file:
    quiz_data = json.load(file)

# Convert the list of questions into a DataFrame
df = pd.DataFrame(quiz_data["questions"])

# Load a pre-trained language model for question analysis (TensorFlow version)
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSequenceClassification.from_pretrained(model_name)

# Create a zero-shot classification pipeline
nlp_pipeline = pipeline("zero-shot-classification", model=model, tokenizer=tokenizer)

# Function to generate contextually relevant links using AI
def generate_smart_link(question):
    # Use the zero-shot classification model to classify the question into categories
    candidate_labels = ["Health", "Biology", "Anatomy", "Medicine", "Science", "Education"]
    result = nlp_pipeline(question, candidate_labels)

    # Choose the most likely label
    predicted_label = result['labels'][0]

    # Generate a link based on the predicted topic
    if predicted_label == "Health":
        base_url = "https://www.healthline.com/search?q="
    elif predicted_label == "Biology":
        base_url = "https://www.khanacademy.org/search?search="
    elif predicted_label == "Anatomy":
        base_url = "https://www.wikipedia.org/wiki/"
    elif predicted_label == "Medicine":
        base_url = "https://www.mayoclinic.org/search/search-results?q="
    else:
        base_url = "https://www.wikipedia.org/wiki/"

    # Format the query to be URL friendly
    query = question.replace(" ", "+")  # Replace spaces with '+' for the URL
    return base_url + query

# Generate contextually relevant links for each question
df["link"] = df["question"].apply(generate_smart_link)

# Display the DataFrame with generated links
df

# # Save the updated DataFrame to a new JSON or CSV file if needed
# df.to_json("quiz_with_smart_links.json", orient="records", indent=4)
```

# Advanced Zero-Shot Classification

```python
import pandas as pd
import json
from transformers import pipeline, AutoTokenizer, TFAutoModelForSequenceClassification

# Load the quiz data from the JSON file
file_path = "quiz.json"
with open(file_path, "r") as file:
    quiz_data = json.load(file)

# Convert the list of questions into a DataFrame
df = pd.DataFrame(quiz_data["questions"])

# Load a pre-trained language model for question analysis (TensorFlow version)
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSequenceClassification.from_pretrained(model_name)

# Create a zero-shot classification pipeline
nlp_pipeline = pipeline("zero-shot-classification", model=model, tokenizer=tokenizer)

# Function to generate contextually relevant links using AI
def generate_smart_link(question):
    # Use the zero-shot classification model to classify the question into categories
    candidate_labels = ["Health", "Biology", "Anatomy", "Medicine", "Science", "Education"]
    result = nlp_pipeline(question, candidate_labels)

    # Choose the most likely label
    predicted_label = result['labels'][0]

    # Generate a link based on the predicted topic
    if predicted_label == "Health":
        base_url = "https://www.healthline.com/search?q="
    elif predicted_label == "Biology":
        base_url = "https://www.khanacademy.org/search?search="
    elif predicted_label == "Anatomy":
        base_url = "https://www.wikipedia.org/wiki/"
    elif predicted_label == "Medicine":
        base_url = "https://www.mayoclinic.org/search/search-results?q="
    else:
        base_url = "https://www.wikipedia.org/wiki/"

    # Format the query to be URL friendly
    query = question.replace(" ", "+")  # Replace spaces with '+' for the URL
    return base_url + query

# Function to enrich questions with additional context
def enrich_question(row):
    explanation_prefix = "This question relates to "
    topic = generate_smart_link(row["question"]).split('/')[2]  # Get the domain for the explanation
    
    # Create an explanation based on the topic
    explanation = (
        explanation_prefix + topic +
        ". The question tests your understanding of key concepts in this area."
    )
    return explanation

# Enrich questions and generate smart links
df["link"] = df["question"].apply(generate_smart_link)
df["explanation"] = df.apply(enrich_question, axis=1)

df
```

# Machine Learning Implementation

```python
import pandas as pd
import json
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the enriched quiz data from finalquiz.json
file_path = "dataset/NLP_Quiz.json"
with open(file_path, "r") as file:
    quiz_data = json.load(file)

# Convert JSON data to a DataFrame
df = pd.DataFrame(quiz_data["questions"])

# Inspect the dataset structure
print("Dataset Sample:\n", df.head())

# Extract features (questions) and labels (categories)
df["category"] = df["explanation"].str.extract(r"relates to (.*?)\.")  # Extract category from explanation

# Fill missing categories using machine learning
missing_categories = df["category"].isna().sum()
print(f"Number of missing categories: {missing_categories}")

if missing_categories > 0:
    # Preprocessing: TF-IDF Vectorization
    vectorizer = TfidfVectorizer(max_features=500)
    features = df["question"]
    X = vectorizer.fit_transform(features).toarray()

    # Encode labels for non-missing categories
    df_non_missing = df.dropna(subset=["category"])
    y_non_missing = pd.factorize(df_non_missing["category"])[0]  # Encode labels

    # Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(
        X[df_non_missing.index], y_non_missing, test_size=0.2, random_state=42
    )

    # Train Random Forest Classifier
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Evaluate Model on Known Data
    y_pred = model.predict(X_test)
    print("\nClassification Report for Known Data:\n")
    print(classification_report(y_test, y_pred))

    # Predict Missing Categories
    df_missing = df[df["category"].isna()]
    missing_predictions = model.predict(X[df_missing.index])
    df.loc[df_missing.index, "category"] = pd.factorize(df_non_missing["category"])[1][missing_predictions]

df
```

```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import pipeline

# Load the enriched quiz data from finalquiz.json
file_path = "dataset/NLP_Quiz.json"
with open(file_path, "r") as file:
    quiz_data = json.load(file)

# Convert JSON data to a DataFrame
df = pd.DataFrame(quiz_data["questions"])

# Inspect the dataset structure
print("Dataset Sample:\n", df.head())

# Load pre-trained model from HuggingFace for zero-shot classification
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define possible candidate labels for classification
candidate_labels = ["Digestive", "Urinary", "Respiratory", "Cardiovascular", "Nervous", "Other"]

# Function to dynamically classify the explanation based on the question and explanation
def dynamic_categorize_system(question, explanation):
    # Combine question and explanation for more accurate classification
    combined_text = f"Question: {question} Explanation: {explanation}"
    
    # Use the classifier to predict the most likely label for the combined text
    result = classifier(combined_text, candidate_labels)
    return result["labels"][0]  # Return the most likely label

# Apply the dynamic categorization function to the question and explanation columns
df["explanation"] = df.apply(lambda row: dynamic_categorize_system(row["question"], row["explanation"]), axis=1)

# Check for missing categories
missing_categories = df["explanation"].isna().sum()
print(f"Number of missing categories: {missing_categories}")

if missing_categories > 0:
    # Combine 'question' and 'explanation' as features for machine learning
    df["combined_text"] = df["question"] + " " + df["explanation"].fillna("")

    # Preprocessing: TF-IDF Vectorization
    vectorizer = TfidfVectorizer(max_features=500, stop_words="english")
    features = vectorizer.fit_transform(df["combined_text"]).toarray()

    # Encode labels for non-missing categories
    df_non_missing = df.dropna(subset=["explanation"])
    y_non_missing = pd.factorize(df_non_missing["explanation"])[0]  # Encode labels

    # Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(
        features[df_non_missing.index], y_non_missing, test_size=0.2, random_state=42
    )

    # Train Random Forest Classifier
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Evaluate Model on Known Data
    y_pred = model.predict(X_test)
    print("\nClassification Report for Known Data:\n")
    print(classification_report(y_test, y_pred))

    # Predict Missing Categories
    df_missing = df[df["explanation"].isna()]
    missing_predictions = model.predict(features[df_missing.index])
    df.loc[df_missing.index, "explanation"] = pd.factorize(df_non_missing["explanation"])[1][missing_predictions]

    # Drop the 'combined_text' column only if it exists
    if 'combined_text' in df.columns:
        df.drop(columns=["combined_text"], inplace=True)

# Check the updated DataFrame
print("\nUpdated DataFrame:\n", df.head())

# Save the updated DataFrame to a json file
output_file_path = "dataset/ML_Quiz.json.json"
df.to_json(output_file_path, orient='records', lines=True)

print(f"Updated data has been saved to {output_file_path}")
```
